```python
import csv
from faker import Faker
import random
from datetime import date, timedelta

fake = Faker('ru_RU')

# Генерация данных для записей
data = []
with open('input.txt','w') as f:
    for _ in range(100000):
        groom_name = fake.first_name_male() + '_' + fake.last_name_male()
        groom_birthdate = fake.date_of_birth(minimum_age=20, maximum_age=40).strftime('%d.%m.%Y')
        
        bride_name = fake.first_name_female() + '_' + fake.last_name_female()
        bride_birthdate = fake.date_of_birth(minimum_age=20, maximum_age=40).strftime('%d.%m.%Y')
        
        marriage_date = (date.today() - timedelta(days=random.randint(365, 3650))).strftime('%d.%m.%Y')
        zags_number = random.randint(101, 999)
        
        f.write(f'{groom_name}\n{groom_birthdate}\n{bride_name}\n{bride_birthdate}\n{marriage_date}\n{zags_number}')
        f.write('\n')

print("База данных успешно сгенерирована в файле 'input.txt'")

```
